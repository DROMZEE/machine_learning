## Quel est l’intérêt de la régression logistique ?

* raisonnement : *toutes choses étant égales par ailleurs*
* La régression logistique est une technique prédictive
* La régression logistique ou modèle logit est un modèle de régression binomiale.
* Elle vise à construire un modèle permettant de prédire / expliquer les valeurs prises par une variable cible qualitative à partir d’un ensemble de variables explicatives quantitatives ou qualitatives (un codage est nécessaire dans ce cas)
* Associer à un vecteur de variables aléatoires ( x 1 , … , x K )  une variable aléatoire binomiale génériquement notée y .
* un cas particulier de modèle linéaire généralisé.
* y est une variable binaire ou une variable qualitative
    * il ne s’agit plus de prédire un nombre réel mais plutôt la classe associée à une observation. 
    * le but recherché est un classifieur
    * les modèles de régression linéaire ne seront plus adaptés puisque les **fonctions linéaires sont définies dans R**.
    
La régression logistique et la régression linéaire appartiennent à la même famille des modèles GLM (Generalized Linear Models) : dans les deux cas on relie un événement à une combinaison linéaire de variables explicatives. 
* Pour la régression linéaire, la variable dépendante suit une loi normale N(µ, s) où µ est une fonction linéaire des variables explicatives. 
* Pour la régression logistique, la variable dépendante, aussi appelée variable réponse, suit une loi de Bernoulli de paramètre p (p la probabilité moyenne pour que l'événement se produise), lorsque l'expérience est répétée une fois, ou une loi Binomiale(n, p) si l'expérience est répétée n fois (par exemple la même dose est essayée sur n insectes). Le paramètre de probabilité p est ici une fonction d'une combinaison linéaire des variables explicatives.


Se distingue de la régression linéaire par lefait que celle-ci demande une distribution normale des variables indépendantes et dépendante. De plus, si on utilisait la régression linéaire, la valeur prédite pourrait se situer en dehors de l’intervalle 0,1.

Se distingue de l’analyse discriminante entrea utres par le fait que celle-ci demande une distribution multi-normale des variables indépendantes.

### Logit vs Probit

Les fonctions les plus couramment utilisées pour relier la probabilité p aux variables explicatives sont la fonction logistique (on parle alors de modèle Logit) et la fonction de répartition de la loi normale standard (on parle alors de modèle Probit). 

Ces deux fonctions sont parfaitement symétriques et [sigmoïdes](https://fr.wikipedia.org/wiki/Sigmo%C3%AFde_(math%C3%A9matiques)).

## Donner des exemples de problèmes auxquels la régression logistique permet de répondre ?

* En médecine, elle permet par exemple de trouver les facteurs qui caractérisent un groupe de sujets malades par rapport à des sujets sains.
* identifier si un email est un spam ou pas
* marketing quantitatif (achat ou non de produits ou services suite à une action)
* finance pour modélisation de risques (scoring).
* dans le domaine phytosanitaire, on cherche à évaluer à partir de quelle dose d'un agent chimique, un insecte sera neutralisé

## Combien y a-t-il de types de régressions logistiques ? Lesquels ?

la variable dépendante est nominale ou ordinale
deux types :
* binaire pour deux catégories.
* polytomique pour plusieurs catégories:  multinominale ou ordinale

### la régression logistique binomiale (ou binaire ou ordinaire)

* la variable d’intérêt est binaire : 0 ou 1, vrai ou faux, oui ou non 
* k = 1
* dummy variables

### la régression logistique multinomiale

Lorsque  la  variable  dépendante  possède  plusieurs  catégories non  ordonnées(K  >  2),  on  parle  de régression logistique multinomiale

* variables cibles qualitatives nominales à plus de 2 modalités,

La   régression   logistique   multinomiale consiste à désigner une catégorie de référence (la  dernière (Kème) par  exemple) pour  fixer  les idées, et à exprimer chaque logit (ou log-odds) des  (K-1)  modalités  par  rapport  à cette référence à l’aide d’une combinaison linéaire des variables prédictives.

### la régression logistique ordinale

* la variable qualitative présente au moins 3 modalités ordonnées hiérarchiquement




> Logit : p = exp(βX) / (1 + exp(βX))

> Probit : p = 1/√2π ∫-∞...βXexp(-x²/2)∂x

> Log-log complémentaire : p = 1 – exp[-exp(βX)]

> Gompertz : p = exp[-exp(βX)]

## Que sont les False Positives et les False Negatives ?

Procédons aux définitions suivantes :

    "Loup" est une classe positive.
    "Pas de loup" est une classe négative.

Nous pouvons résumer notre modèle de prédiction de loup avec une matrice de confusion 2x2 qui illustre les quatre résultats possibles :

Vrai positif (VP) : | Faux positif (FP) :
----------------------|-------------------------------
Réalité : attaque de loup.| Réalité : pas d'attaque de loup.
Le berger dit : "Loup".|Le berger dit : "Loup".
Résultat : le berger est un héros.|Résultat : les villageois en veulent au berger de les avoir réveillés.


Faux négatif (FN) : | Vrai négatif (VN) :
---------------------------|----------------------
Réalité : attaque de loup.|Réalité : pas d'attaque de loup.
Le berger dit : "Pas de loup".|Le berger dit : "Pas de loup".
Résultat : le loup dévore tous les moutons.| Résultat : tout le monde est sain et sauf.


Un vrai positif est un résultat où le modèle prédit correctement la classe positive. De façon analogue, un vrai négatif est un résultat où le modèle prédit correctement la classe négative.

Un faux positif est un résultat où le modèle prédit incorrectement la classe positive. Et un faux négatif est un résultat où le modèle prédit incorrectement la classe négative.

Le taux de faux positifs (TFP) : proportion de négatifs qui ont été classés positifs

## Que sont les notions de :

### performance ou justesse ou accuracy,


* χ2 du modèle: Est-ce que l’ensemble de variables – ou les variables du bloc entré – est significativement reliéà la VD?
* χ2 de maximum de vraisemblance (Hosmer et Lemeshow): est-ce que les données sont compatibles avec les relations postulées?
* Classification: Jusqu’à quel point les informations sur les VI permettent de bien classer les cas dans les catégories de la VD, étant donné le point de coupure.
* Analyse des résidus: proportion de résidus à plus de 3 écarts-types (si n est grand), distribution normale des résidus.


**maximum de vraisemblance**

* Estimateurs des paramètres sans biais et de faible variance
* La vraisemblance d’un n-échantillon y1,y2,...,ynest définie comme la probabilité d’observer cet échantillon


### d’erreur,

Taux d’erreur : nombre de mauvais classement rapporté à l’effectif total

Taux de succès : probabilité de bon classement du modèle

### de précision ou valeur prédictive positive,

proportion de vrais positifs parmi les individus classés positifs

La F-Mesure: moyenne harmonique entre la sensibilité et la précision


### de sensibilité ou rappel ou recall,

Sensibilité : capacité du modèle à retrouver les positifs

### de spécificité ?

Spécificité : proportion de négatifs détectés

## Comment interprète-t-on les coefficients de la régression logistique ?

* Risque relatif : surcroit de chance d’être positif du groupe exposé par rapport au groupe témoin
* Odds : rapport de probabilités dans un groupe
* Odds-ratio : rapport entre l’odds du groupe exposé et l’odds du groupe témoin
* La statistique de WALD



## Que sont les odds ratio (rapport de côtes) ?

Il s’agit d’une transformation logarithmique, difficile àinterpréter. C’est pourquoi on retransforme parl’inverse (Exp) pour faciliter l’interprétation. On parlealors de rapport de cote (odds ratio).

C’est le rapport des cotes des probabilités d’avoir la maladie pour ceux qui ont un symptôme X d’une part et de ceux qui ne l’ont pas d’autre part.
* OR=1, la maladie est indépendantedu symptôme
* OR>1, la maladie est plus fréquente pour les individus qui ontle symptôme.
* OR<1, la maladie est plus fréquente pour les individus qui n’ont pas le symptôme.

## Que sont les critères ROC et AUC ?

Une courbe ROC (receiver operating characteristic) est un graphique représentant les performances d'un modèle de classification pour tous les seuils de classification. Cette courbe trace le taux de vrais positifs en fonction du taux de faux positifs

AUC signifie "aire sous la courbe ROC". Cette valeur mesure l'intégralité de l'aire à deux dimensions située sous l'ensemble de la courbe ROC (par calculs d'intégrales) de (0,0) à (1,1).
